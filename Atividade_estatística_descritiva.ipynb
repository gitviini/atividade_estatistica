## 3. Análise de Quartis e Outliers

Os quartis dividem um dataset ordenado em quatro partes iguais (Q1: 25%, Q2: 50% - mediana, Q3: 75%). Eles são cruciais para entender a dispersão e identificar possíveis outliers. Um outlier é um ponto de dado que se desvia significativamente de outras observações. Podemos identificá-los usando o Intervalo Interquartil (IQR), que é a diferença entre Q3 e Q1, e estabelecendo limites inferior e superior.

### Exercício 3.1: Cálculo Manual de Quartis e Identificação de Outliers

Para a coluna 'total_bill', calcule manualmente (sem usar `describe()` diretamente para os quartis, mas pode usar `quantile()`) os valores de Q1, Q2 (mediana), Q3 e o Intervalo Interquartil (IQR). Em seguida, calcule o Limite Inferior e Superior para identificação de outliers (Limite Inferior = Q1 - 1.5 * IQR; Limite Superior = Q3 + 1.5 * IQR). Por fim, mostre quantos outliers existem na coluna 'total_bill'.

```python
# DIGITE SEU CÓDIGO AQUI

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis


df = sns.load_dataset('tips')

print(df.describe())

print("=== Total Bill ===")
print("Q1 = ", df['total_bill'].quantile(0.25))
print("Q2 = ", df['total_bill'].quantile(0.5))
print("Q3 = ", df['total_bill'].quantile(0.75))

q1_total_bill = df['total_bill'].quantile(0.25)
q3_total_bill = df['total_bill'].quantile(0.75)

iqr_total_bill = q3_total_bill - q1_total_bill

print("IQR = ", iqr_total_bill)

limite_superior_total_bill = q3_total_bill + 1.5 * iqr_total_bill
limite_inferior_total_bill = q1_total_bill - 1.5 * iqr_total_bill

print("=== Intervalo Interquartil ===")
print("Limite Inferior = ", limite_inferior_total_bill)
print("Limite Superior = ", limite_superior_total_bill)

outliers_total_bill = df[(df['total_bill'] < limite_inferior_total_bill) | (df['total_bill'] > limite_superior_total_bill)]
print("Quantidade de outliers:", len(outliers_total_bill))

print("=== Tip ===")
print("Q1 = ", df['tip'].quantile(0.25))
print("Q2 = ", df['tip'].quantile(0.5))
print("Q3 = ", df['tip'].quantile(0.75))

q1_tip = df['tip'].quantile(0.25)
q3_tip = df['tip'].quantile(0.75)

iqr_tip = q3_tip - q1_tip

print("IQR = ", iqr_tip)

limite_superior_tip = q3_tip + 1.5 * iqr_tip
limite_inferior_tip = q1_tip - 1.5 * iqr_tip

print("=== Intervalo Interquartil ===")
print("Limite Inferior = ", limite_inferior_tip)
print("Limite Superior = ", limite_superior_tip)

outliers_tip = df[(df['tip'] < limite_inferior_tip) | (df['tip'] > limite_superior_tip)]
print("Quantidade de outliers:", len(outliers_tip))

print("=== Size ===")
print("Q1 = ", df['size'].quantile(0.25))
print("Q2 = ", df['size'].quantile(0.5))
print("Q3 = ", df['size'].quantile(0.75))

q1_size = df['size'].quantile(0.25)
q3_size = df['size'].quantile(0.75)

iqr_size = q3_size - q1_size

print("IQR = ", iqr_size)

limite_superior_size = q3_size + 1.5 * iqr_size
limite_inferior_size = q1_size - 1.5 * iqr_size

print("=== Intervalo Interquartil ===")
print("Limite Inferior = ", limite_inferior_size)
print("Limite Superior = ", limite_superior_size)
outliers_size = df[(df['size'] < limite_inferior_size) | (df['size'] > limite_superior_size)]
print("Quantidade de outliers:", len(outliers_size))
```

### Análise do Aluno:
Foram encontrados 9 outliers, o IQR de 10.78 representa a amplitude central da faixa, ou seja, os 50% dos clientes desse restaurante gastam numa faixa de $10.78 de variação. Na análise do contexto atual, de um restaurante, a investigação desses valores extremos facilitaria na melhoria da administração de um negócio. ## 4. Assimetria (Skewness) e Curtose

A assimetria (skewness) mede o grau de inclinação de uma distribuição de probabilidade. Uma distribuição simétrica tem skewness zero. Valores positivos indicam uma cauda direita mais longa, enquanto valores negativos indicam uma cauda esquerda mais longa.

A curtose (kurtosis) mede o 'achatamento' ou 'pico' de uma distribuição. Uma curtose alta significa mais valores nas caudas e um pico mais nítido (leptocúrtica), enquanto uma curtose baixa indica menos valores nas caudas e um pico mais plano (platocúrtica). Uma distribuição normal tem curtose próxima de 3 (ou 0, se for calculada a curtose em excesso).

## 7. Normalização de Dados

A normalização é um processo crucial no pré-processamento de dados, especialmente quando algoritmos de Machine Learning são sensíveis à escala das features (por exemplo, modelos baseados em distância como KMeans, SVMs com kernels, ou Redes Neurais). O objetivo é ajustar os valores das features para uma escala comum, sem distorcer as diferenças nas faixas de valores ou perder informações. Existem duas técnicas comuns:

1.  **Min-Max Scaling (Normalização)**: Escala os dados para um intervalo fixo, geralmente entre 0 e 1. A fórmula é: `X_novo = (X - X_min) / (X_max - X_min)`.
2.  **StandardScaler (Padronização - Z-score)**: Transforma os dados para ter média 0 e desvio padrão 1. A fórmula é: `X_novo = (X - média) / desvio_padrão`.

### Exercício 7.1: Aplicação de Normalização Min-Max e StandardScaler

Utilize as colunas numéricas ('total_bill', 'tip', 'size') para aplicar dois métodos de normalização:

1.  **Min-Max Scaling**: Normalize estas colunas para o intervalo [0, 1]. Armazene os resultados em um novo DataFrame ou adicione colunas normalizadas ao DataFrame `df` (ex: `total_bill_minmax`).
2.  **StandardScaler (Z-score)**: Padronize as mesmas colunas para ter média 0 e desvio padrão 1. Armazene os resultados em um novo DataFrame ou adicione colunas padronizadas ao DataFrame `df` (ex: `total_bill_zscore`).

Lembre-se de importar `MinMaxScaler` e `StandardScaler` do `sklearn.preprocessing`.

```python
# DIGITE SEU CÓDIGO AQUI

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
from sklearn.preprocessing import MinMaxScaler, StandardScaler


df = sns.load_dataset('tips')

#Selecionando as colunas numéricas para a normalização

colunas = df[['total_bill', 'tip', 'size']]

#Min-Max Scaler

scaler_minmax = MinMaxScaler()

resultado_minmax = scaler_minmax.fit_transform(colunas)

df[['total_bill_minmax', 'tip_minmax', 'size_minmax']] = resultado_minmax

#Z-Score Scaler

scaler_z_score = StandardScaler()
resultado_z_score = scaler_z_score.fit_transform(colunas)
df[['total_bill_z_score', 'tip_z_score', 'size_z_score']] = resultado_z_score

#Print para verificar as saídas

print(df[['total_bill', 'total_bill_minmax', 'total_bill_z_score']].head())
print("\n")
print(df[['tip', 'tip_minmax', 'tip_z_score']].head())
print("\n")
print(df[['size', 'size_minmax', 'size_z_score']].head())
```

### Análise do Aluno:
A normalização Min-Max gera valores entre 0 e 1, sendo indicada quando os dados possuem um intervalo conhecido e sem outliers severos. No contexto analisado, a técnica se aplica melhor à variável tips, que possui um intervalo fixo e específico entre 1 e 6, não podendo apresentar valores além dos que se encontram entre estes dois. Para as variáveis restantes, tip e total_bill, a normalização não seria adequada, por possuírem outliers acima do limite superior, distorcendo a escala. A normalização Z-Score é ideal quando o objetivo é comparar variáveis com escalas diferentes, centralizando os dados em média 0 e desvio padrão em 1. Apesar de ser mais robusto que o Min-Max, ainda é uma técnica que causa distorções na escala por usar a média como centro. A técnica ideal para a coluna total_bill seria o Robust Scaler, baseado no IQR (Intervalo Interquartil), por se tratar de uma técnica mais robusta que utiliza a mediana como centro e o IQR como escala, sendo mais resistente aos outliers e representando melhor a distribuição real dos dados.
